```python
#constants using iharmony's units
m = 9.1093826e-28
c = 2.99792458e10
theta_e = 10
e = 4.80320680e-10
B = 30.
h = 6.6260693e-27
#n_e = 1.e5
nu_c = (e * B)/(2 * np.pi * m * c)
#nuratio = np.arange(1., 1.e6, 1)

#variables to be set (frequency, observer angle, width of delta function, electron number density, 
# sum index n)
#nu = 5. * nu_c
theta = 12. * np.pi / 36.
n_e = 1.
nu_s = (2./9.) * nu_c * theta_e**2. * np.sin(theta)

n_e_NT = 1.
p = 3.
gamma_min = 1.
gamma_max = 1000.
```

```python
def eq_58(nu_ratio):
    prefactor = n_e_NT*(e**2.*nu_c)/c*(3.**(p/2.)*(p-1.)*np.sin(theta))/(2.*(p+1.)*(gamma_min**(1.-p)-gamma_max**(1.-p)))
    gamma_fns = special.gamma((3. * p - 1.)/12.) * special.gamma((3. * p + 19.)/12.)
    ans = prefactor * gamma_fns * (nu_ratio/(np.sin(theta)))**(-(p - 1.)/2.)
    return ans

def eq_58_q(nu_ratio):
    ans = eq_58(nu_ratio) * ((p+1.)/(p+7./3.))
    return ans 

def eq_58_v(nu_ratio):
    ans = eq_58(nu_ratio) * 171./250. * np.sqrt(p)/np.tan(theta) * nu_ratio**(-1./2.) * np.sqrt(3.*np.sin(theta))
    return ans
```

```python
def dexter_pl_I(nu_ratio):
    nu = nu_ratio * nu_c
    nu_p = (3./2. * nu_c * np.sin(theta))
    prefactor = n_e_NT * e**2. * (p-1.)* nu_p / (2. * np.sqrt(3.)* c * (gamma_min**(1.-p)-gamma_max**(1.-p)))
    mid_term = (nu/nu_p)**(-(p-1.)/2.) * 2.**((p-3.)/2.) * (p+7./3.)/(p+1.)
    gamma_fns = special.gamma(p/4. + 7./12.) * special.gamma(p/4. - 1./12.)
    ans = prefactor * mid_term * gamma_fns
    return ans

def dexter_pl_Q(nu_ratio):
    ans = (p+1.)/(p+7./3.) * dexter_pl_I(nu_ratio)
    return ans
    
def dexter_pl_V(nu_ratio):
    nu = nu_ratio * nu_c
    nu_p = (3./2. * nu_c * np.sin(theta))
    prefactor = 2. * n_e_NT * e**2. * (p-1.) * nu_p * 1./np.tan(theta) / (3. * np.sqrt(3.) * c * (gamma_min**(1.-p)-gamma_max**(1.-p)))
    mid_term = (nu/nu_p)**(-p/2.) * 2.**(p/2. - 1.) * (p+2.)/p
    gamma_fns = special.gamma(p/4. + 1./3.) * special.gamma(p/4. + 2./3.)
    ans = prefactor * mid_term * gamma_fns
    return ans

nu_ratio = np.arange(1., 1000., 0.1)

#test for Stokes I
#pl.loglog(nu_ratio, dexter_pl_I(nu_ratio))
#pl.loglog(nu_ratio, eq_58(nu_ratio))

#test for Stokes Q
#pl.loglog(nu_ratio, dexter_pl_Q(nu_ratio))
#pl.loglog(nu_ratio, eq_58_q(nu_ratio))

#test for Stokes V
pl.loglog(nu_ratio, dexter_pl_V(nu_ratio))
pl.loglog(nu_ratio, eq_58_v(nu_ratio))
```

```python
#PL absorptivity tests

def eq_59_I(nu_ratio):
    nu = nu_ratio * nu_c
    prefactor = n_e_NT * (e**2./(nu*m*c)) * 3.**((p+1.)/2.) * (p-1.)/(4.*(gamma_min**(1.-p)-gamma_max**(1.-p)))
    gamma_fns = special.gamma((3.*p+2.)/12.) * special.gamma((3.*p+22.)/12.)
    ans = prefactor * gamma_fns * (nu/(nu_c * np.sin(theta)))**(-(p+2.)/2.)
    return ans

def dexter_abs_I(nu_ratio):
    nu = nu_ratio * nu_c
    nu_p = (3./2. * nu_c * np.sin(theta))
    prefactor_top = n_e_NT * e**2. * (p-1.)*(p+2.)
    prefactor_bot = 4.*np.sqrt(3.)*m*c*nu_p*(gamma_min**(1.-p)-gamma_max**(1.-p))
    mid = (nu/nu_p)**(-p/2. - 2.) * (p+10./3.)/(p+2.) * 2.**(p/2.-1.) #error in print here (p+1)->(p+2)
    gamma_fns = special.gamma(p/4. + 5./6.) * special.gamma(p/4. + 1./6.)
    ans = prefactor_top/prefactor_bot * mid * gamma_fns
    return ans

def our_abs_Q(nu_ratio):
    ans = eq_59_I(nu_ratio) * 3./4. * (p-1.)**(43./500.)
    return ans

def dexter_abs_Q(nu_ratio):
    ans = (p+2.)/(p+10./3.) * dexter_abs_I(nu_ratio)
    return ans

def our_abs_V(nu_ratio):
    nu = nu_ratio * nu_c
    p1 = 7./4. * (71.*p/100. + 22./625.)**(197./500.) * (np.sin(theta)**(-48./25.)-1.)**(64./125.)
    p2 = (nu/(nu_c * np.sin(theta)))**(-1./2.)
    ans = eq_59_I(nu_ratio) * p1 * p2
    return ans

def dexter_abs_V(nu_ratio):
    nu = nu_ratio * nu_c
    nu_p = (3./2. * nu_c * np.sin(theta))
    prefactor_top = n_e_NT * e**2. * (p-1.) * (p+2.) * 1./np.tan(theta)
    prefactor_bot = 3. * np.sqrt(3.) * m * c * nu_p * (gamma_min**(1.-p)-gamma_max**(1.-p))
    mid = (nu/nu_p)**(-(p+5.)/2.) * (p+3.)/(p+1.) * 2.**((p-1.)/2.)
    gamma_fns = special.gamma(p/4. + 7./12.) * special.gamma(p/4. + 11./12.)
    ans = prefactor_top/prefactor_bot * mid * gamma_fns
    return ans


#Stokes I abs tests
#pl.loglog(nu_ratio, dexter_abs_I(nu_ratio))
#pl.loglog(nu_ratio, eq_59_I(nu_ratio))

#Stokes Q abs tests
#pl.loglog(nu_ratio, dexter_abs_Q(nu_ratio))
#pl.loglog(nu_ratio, our_abs_Q(nu_ratio))

#Stokes V abs tests
pl.loglog(nu_ratio, dexter_abs_V(nu_ratio))
pl.loglog(nu_ratio, our_abs_V(nu_ratio))
```
