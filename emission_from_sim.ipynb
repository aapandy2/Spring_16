```python
#constants using iharmony's units
m = 9.1093826e-28
c = 2.99792458e10
theta_e = 10.
e = 4.80320680e-10
h = 6.6260693e-27

gamma_min = 1.
gamma_max = 100.
p = 3.5

kappa = 3.5
w = 10.

nuratio = 1.e2
```

```python
#emissivity fitting formulae
def thermal_I(nuratio, obs_angle, B_mag, density):
    nu_c = (e * B_mag)/(2 * np.pi * m * c)
    nu_s = (2./9.) * nu_c * theta_e**2. * np.sin(obs_angle)
    term1 = density * (np.sqrt(2.) * np.pi * e**2. * nu_s)/(3. * special.kn(2., 1./theta_e) * c)
    nu = nuratio * nu_c
    X = nu/nu_s
    term2 = (X**(1./2.) + 2.**(11./12.) * X**(1./6.))**2. * np.exp(-X**(1./3.))
    return term1*term2

def thermal_Q(nuratio, obs_angle, B_mag, density):
    nu_c = (e * B_mag)/(2 * np.pi * m * c)
    nu_s = (2./9.) * nu_c * theta_e**2. * np.sin(obs_angle)
    nu = nuratio * nu_c
    X = nu/nu_s
    prefactor = density * e**2. * nu_c / c
    A = (7.*theta_e**(24./25.)+35.)/(10.*theta_e**(24./25.)+75.)
    ans = (2.**(1./2.)*np.pi/27.) * np.sin(obs_angle) * (X**(1./2.) + A*2.**(11./12.)*X**(1./6.))**2. * np.exp(-X**(1./3.))
    return prefactor * ans

def thermal_V(nuratio, obs_angle, B_mag, density):
    nu_c = (e * B_mag)/(2 * np.pi * m * c)
    nu_s = (2./9.) * nu_c * theta_e**2. * np.sin(obs_angle)
    nu = nuratio * nu_c
    X = nu/nu_s
    prefactor = density * e**2. * nu_c / c * np.exp(-X**(1./3.))
    ans = (37.-87.*np.sin(obs_angle-28./25.))/(100.*(theta_e+1.)) * (1.+(theta_e**(3./5.) /25.+7./10.)*X**(9./25.))**(5./3.)
    return prefactor * ans

def pl_I(nuratio, obs_angle, B_mag, density):
    nu_c = (e * B_mag)/(2 * np.pi * m * c)
    prefactor = density*(e**2.*nu_c)/c*(3.**(p/2.)*(p-1.)*np.sin(obs_angle))/(2.*(p+1.)*(gamma_min**(1.-p)-gamma_max**(1.-p)))
    gamma_fns = special.gamma((3. * p - 1.)/12.) * special.gamma((3. * p + 19.)/12.)
    ans = prefactor * gamma_fns * (nuratio/(np.sin(obs_angle)))**(-(p - 1.)/2.)
    return ans

def pl_Q(nuratio, obs_angle, B_mag, density):
    ans = pl_I(nuratio, obs_angle, B_mag, density) * ((p+1.)/(p+7./3.))
    return ans 

def pl_V(nuratio, obs_angle, B_mag, density):
    nu_c = (e * B_mag)/(2 * np.pi * m * c)
    nu = nuratio * nu_c
    a = (171./250.)*p**(49./100.)
    ans = pl_I(nuratio, obs_angle, B_mag, density) * a/(np.tan(obs_angle)) * (nu/(3.*nu_c*np.sin(obs_angle)))**(-1./2.)
    return ans

def kappa_I(nuratio, obs_angle, B_mag, density):
    nu_c = (e * B_mag)/(2 * np.pi * m * c)
    nu = nuratio * nu_c
    prefactor = density * e**2. * nu_c / c
    nu_w = nu_c * (w*kappa)**2. * np.sin(obs_angle)
    X_kappa = nu / nu_w
    low = X_kappa**(1./3.) * np.sin(obs_angle)*(4.*np.pi*special.gamma(kappa-4./3.))/(3.**(7./3.)*special.gamma(kappa-2.))
    high1 = X_kappa**(-(kappa-2.)/2.) * np.sin(obs_angle) * 3.**((kappa-1.)/2.) * (kappa-2.)*(kappa-1.)/4.
    high2 = special.gamma(kappa/4.-1./3.) * special.gamma(kappa/4. + 4./3.)
    x = 3.*kappa**(-3./2.)
    ans = prefactor * (low**(-x) + (high1*high2)**(-x))**(-1./x)
    return ans

def kappa_Q(nuratio, obs_angle, B_mag, density):
    nu_c = (e * B_mag)/(2 * np.pi * m * c)
    nu = nuratio * nu_c
    prefactor = density * e**2. * nu_c / c
    nu_w = nu_c * (w*kappa)**2. * np.sin(obs_angle)
    X_kappa = nu / nu_w
    low = X_kappa**(1./3.) * np.sin(obs_angle)*(4.*np.pi*special.gamma(kappa-4./3.))/(3.**(7./3.)*special.gamma(kappa-2.))
    high1 = X_kappa**(-(kappa-2.)/2.) * np.sin(obs_angle) * 3.**((kappa-1.)/2.) * (kappa-2.)*(kappa-1.)/4.
    high2 = special.gamma(kappa/4.-1./3.) * special.gamma(kappa/4. + 4./3.)
    xQ = 37./10. * kappa**(-8./5.)
    lowQ = low * (1./2.)
    highQ = high1 * high2 * ((4./5.)**2. + kappa/50.)
    ans = prefactor * (lowQ**(-xQ) + highQ**(-xQ))**(-1./xQ)
    return -ans

def kappa_V(nuratio, obs_angle, B_mag, density):
    nu_c = (e * B_mag)/(2 * np.pi * m * c)
    nu = nuratio * nu_c
    prefactor = density * e**2. * nu_c / c
    nu_w = nu_c * (w*kappa)**2. * np.sin(obs_angle)
    X_kappa = nu / nu_w
    low = X_kappa**(1./3.)*np.sin(obs_angle)*(4.*np.pi*special.gamma(kappa-4./3.))/(3.**(7./3.)*special.gamma(kappa-2.))
    high1 = X_kappa**(-(kappa-2.)/2.) * np.sin(obs_angle) * 3.**((kappa-1.)/2.) * (kappa-2.)*(kappa-1.)/4.
    high2 = special.gamma(kappa/4.-1./3.) * special.gamma(kappa/4. + 4./3.)
    xV = 13./5. * kappa**(-36./25.)
    lowV = low *(3./4.)**2.*(np.sin(obs_angle)**(-12./5.)-1.)**(12./25.)*kappa**(-66./125.)/w*X_kappa**(-7./20.)
    highV = high1*high2*(7./8.)**2. * (np.sin(obs_angle)**(-5./2.)-1.)**(11./25.)*kappa**(-11./25.)/w*X_kappa**(-1./2.)
    ans = prefactor * (lowV**(-xV) + highV**(-xV))**(-1./xV)
    return -ans
    
```

```python
#absorptivity fitting formulae

def planck_func(nuratio, B_mag):
    nu_c = (e * B_mag)/(2 * np.pi * m * c)
    nu = nuratio * nu_c
    term1 = (2.*h*nu**3.)/c**2.
    term2 = (np.exp(h*nu/(theta_e*m*c**2.))-1.);
    ans = term1 / term2;
    return ans;

def thermal_I_abs(nuratio, obs_angle, B_mag, density):
    ans = thermal_I(nuratio, obs_angle, B_mag, density)/planck_func(nuratio, B_mag)
    return ans

def thermal_Q_abs(nuratio, obs_angle, B_mag, density):
    ans = thermal_Q(nuratio, obs_angle, B_mag, density)/planck_func(nuratio, B_mag)
    return ans

def thermal_V_abs(nuratio, obs_angle, B_mag, density):
    ans = thermal_V(nuratio, obs_angle, B_mag, density)/planck_func(nuratio, B_mag)
    return ans

def pl_I_abs(nuratio, obs_angle, B_mag, density):
    nu_c = (e * B_mag)/(2 * np.pi * m * c)
    nu = nuratio * nu_c
    prefactor = density * (e**2.)/(nu*m*c) * (3.**((p+1.)/2.) * (p-1.))/(4.*(gamma_min**(1.-p)-gamma_max**(1.-p)))
    gamma_fns = special.gamma((3.*p+2.)/12.) * special.gamma((3.*p+22.)/12.)
    ans = prefactor * gamma_fns * (nuratio/np.sin(obs_angle))**(-(p+2.)/2.)
    return ans

def pl_Q_abs(nuratio, obs_angle, B_mag, density):
    nu_c = (e * B_mag)/(2 * np.pi * m * c)
    nu = nuratio * nu_c
    prefactor = (density*e**2.)/(nu*m*c);
    term1 = 3.**((p+1.)/2.)*(p-1.)
    term2 = 4.*(gamma_min**(1.-p)-gamma_max**(1.-p))
    term3 = special.gamma((3.*p+2.)/12.)*special.gamma((3.*p+22.)/12.);
    term4 = (nu/(nu_c*np.sin(obs_angle)))**(-(p+2.)/2.)
    term5 = -((17./500.)*p - 43./1250.)**(43./500.)
    ans = prefactor*term1/term2*term3*term4*term5
    return ans

def pl_V_abs(nuratio, obs_angle, B_mag, density):
    nu_c = (e * B_mag)/(2 * np.pi * m * c)
    nu = nuratio * nu_c
    prefactor = (density*e**2.)/(nu*m*c)
    term1 = 3.**((p+1.)/2.)*(p-1.)
    term2 = 4.*(gamma_min**(1.-p)-gamma_max**(1.-p))
    term3 = special.gamma((3.*p+2.)/12.)*special.gamma((3.*p+22.)/12.)
    term4 = (nu/(nu_c*np.sin(obs_angle)))**(-(p+2.)/2.)
    term5 = -((71./100.)*p+22./625.)**(197./500.)
    term6 = ((31./10.)*(np.sin(obs_angle))**(-48./25)-31./10.)**(64./125.)
    term7 = (nu/(nu_c*np.sin(obs_angle)))**(-1./2.)
    ans = prefactor*term1/term2*term3*term4*term5*term6*term7
    return ans

def kappa_I_abs(nuratio, obs_angle, B_mag, density):
    nu_c = (e * B_mag)/(2 * np.pi * m * c)
    nu = nuratio * nu_c
    nu_w = (w*kappa)**(2.)*nu_c*np.sin(obs_angle)
    X_k = nu/nu_w
    prefactor = density*e/(B_mag*np.sin(obs_angle))
    a = kappa - 1./3. 
    b = kappa + 1.
    d = kappa + 2./3.
    z = -kappa*w
    hyp2f1 = (1.-z)**(-a)*special.gamma(d)*special.gamma(b-a)/(special.gamma(b)*special.gamma(d-a))*special.hyp2f1(a, d-b, a-b+1., 1./(1.-z))+(1.-z)**(-b)*special.gamma(d)*special.gamma(a-b)/(special.gamma(a)*special.gamma(d-b))*special.hyp2f1(b, d-a, b-a+1., 1./(1.-z))
    Nlow = (3.)**(1./6.)*(10./41.)*(2.*np.pi)**(2.)/(w*kappa)**(16./3.-kappa)*(kappa-2.)*(kappa-1.)*kappa/(3.*kappa-1.)*special.gamma(5./3.)*hyp2f1
    Nhigh = 2.*(np.pi)**(5./2.)/3.*(kappa-2.)*(kappa-1.)*kappa/(w*kappa)**(5.)*(2*special.gamma(2.+kappa/2.)/(2.+kappa)-1.)*((3./kappa)**(19./4.)+3./5.)
    x = (-7./4. + 8.*kappa/5.)**(-43./50.)
    ans = prefactor*Nlow*(X_k)**(-5./3.)*(1.+(X_k)**(x*(3.*kappa-1.)/6.)*(Nlow/Nhigh)**(x))**(-1./x)
    return ans

def kappa_Q_abs(nuratio, obs_angle, B_mag, density):
    nu_c = (e * B_mag)/(2 * np.pi * m * c)
    nu = nuratio * nu_c
    nu_w = (w*kappa)**(2.)*nu_c*np.sin(obs_angle)
    X_k = nu/nu_w
    prefactor = density*e/(B_mag*np.sin(obs_angle))
    a = kappa - 1./3. 
    b = kappa + 1.
    d = kappa + 2./3.
    z = -kappa*w
    hyp2f1 = (1.-z)**(-a)*special.gamma(d)*special.gamma(b-a)/(special.gamma(b)*special.gamma(d-a))*special.hyp2f1(a, d-b, a-b+1., 1./(1.-z))+(1.-z)**(-b)*special.gamma(d)*special.gamma(a-b)/(special.gamma(a)*special.gamma(d-b))*special.hyp2f1(b, d-a, b-a+1., 1./(1.-z))
    Nlow = -(25./48.)*(3.)**(1./6.)*(10./41.)*(2.*np.pi)**(2.)/(w*kappa)**(16./3.-kappa)*(kappa-2.)*(kappa-1.)*kappa/(3.*kappa-1.)*special.gamma(5./3.)*hyp2f1
    Nhigh = -((21.)**(2.)*(kappa)**(-144./25.)+11./20.)*2.*(np.pi)**(5./2.)/3.*(kappa-2.)*(kappa-1.)*kappa/(w*kappa)**(5.)*(2*special.gamma(2.+kappa/2.)/(2.+kappa)-1.)
    x = (7./5.)*(kappa)**(-23./20.)
    ans = prefactor*Nlow*(X_k)**(-5./3.)*(1.+(X_k)**(x*(3.*kappa-1.)/6.)*(Nlow/Nhigh)**(x))**(-1./x)
    return ans

def kappa_V_abs(nuratio, obs_angle, B_mag, density):
    nu_c = (e * B_mag)/(2 * np.pi * m * c)
    nu = nuratio * nu_c
    nu_w = (w*kappa)**(2.)*nu_c*np.sin(obs_angle)
    X_k = nu/nu_w
    prefactor = density*e/(B_mag*np.sin(obs_angle))
    a = kappa - 1./3. 
    b = kappa + 1.
    d = kappa + 2./3.
    z = -kappa*w
    hyp2f1 = (1.-z)**(-a)*special.gamma(d)*special.gamma(b-a)/(special.gamma(b)*special.gamma(d-a))*special.hyp2f1(a, d-b, a-b+1., 1./(1.-z))+(1.-z)**(-b)*special.gamma(d)*special.gamma(a-b)/(special.gamma(a)*special.gamma(d-b))*special.hyp2f1(b, d-a, b-a+1., 1./(1.-z))
    Nlow = -(77./(100.*w))*((np.sin(obs_angle))**(-114./50.)-1.)**(223./500.)*(X_k)**(-7./20.)*(kappa)**(-7./10)*(3.)**(1./6.)*(10./41.)*(2.*np.pi)**(2.)/(w*kappa)**(16./3.-kappa)*(kappa-2.)*(kappa-1.)*kappa/(3.*kappa-1.)*special.gamma(5./3.)*hyp2f1
    Nhigh = -(143./10. * (w)**(-116./125.))*((np.sin(obs_angle))**(-41./20.)-1.)**(1./2.)*(13.*13.*(kappa)**(-8.)+13./(2500.)*kappa - 263./5000.+47./(200.*kappa))*(X_k)**(-1./2.)*2.*(np.pi)**(5./2.)/3.*(kappa-2.)*(kappa-1.)*kappa/(w*kappa)**(5.)*(2*special.gamma(2.+kappa/2.)/(2.+kappa)-1.)
    x = (61./50.)*(kappa)**(-142./125.)+7./1000.
    ans = prefactor*Nlow*(X_k)**(-5./3.)*(1.+(X_k)**(x*(3.*kappa-1.)/6.)*(Nlow/Nhigh)**(x))**(-1./x)
    return ans
```
