```python
obs_vector = [1, 1, 0] #choose observer vector; does not have to be normalized (normalization taken care of below)

obs_angle  = np.zeros(np.shape(B_x))

for i in range(0, np.shape(B_x)[0]):
    for j in range(0, np.shape(B_x)[1]):
        B_vector = [B_x[j][i], B_y[j][i], B_z[j][i]]
        obs_angle[j][i] = np.arccos(np.dot(B_vector, obs_vector) / (np.linalg.norm(B_vector) * np.linalg.norm(obs_vector)))
```

```python
def j_nu_pl_3(n_e, B, obs_angle, nu):
    return n_e * B**2. * np.sin(obs_angle)**2. / nu     #j_nu PL_3 (low error)
#    return n_e * B * np.sin(obs_angle)**2.

def alpha_nu_pl_2(n_e, B, obs_angle, nu):
    return n_e * B**2. * np.sin(obs_angle)**2. / nu**2. #alpha_nu PL_2 (high error)
#    return n_e * 1. * np.sin(obs_angle)**2. / B 

def exact_j(nu):
    return np.mean(j_nu_pl_3(n_e, B_mag, obs_angle, nu))

def avg_j(nu):
    return j_nu_pl_3(np.mean(n_e), np.mean(B_mag), np.mean(obs_angle), nu)

def exact_a(nu):
    return np.mean(alpha_nu_pl_2(n_e, B_mag, obs_angle, nu))

def avg_a(nu):
    return alpha_nu_pl_2(np.mean(n_e), np.mean(B_mag), np.mean(obs_angle), nu)


def error(exact, avg):
    return (exact - avg)/exact
```
